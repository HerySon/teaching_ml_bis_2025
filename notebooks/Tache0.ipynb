# 3°) Filtrage des Colonnes (Réduction de dimensionnalité)

### 3.1°) Filtrage en fonction des valeurs manquantes

Critères pour choisir les colonnes à garder : 
- Pertinence : Garder les colonnes essentielles (ex: nom du produit, marque, code-barres, nutrition, additifs, labels)
- Taux de remplissage : Eliminer les colonnes où +70% des valeurs sont NaN.
- Variabilité : Exclure les colonnes avec une seule valeur unique

```python
#df.notna().mean()
```

```python
# Filtrer votre échantillon avec une certaine tolérance concernant les valeurs non-nulles par colonne
# Exemple
# Eliminer les colonnes où +70% des valeurs sont nulles (NaN)
threshold = 0.3  # Seuil minimal de 30% de valeurs non-nulles
df_filtered = df.loc[:, df.notna().mean() > threshold]
df_filtered
```

```python
## Permet de vérifier que le résultat attendu est bien celui qu'on obtient
# Récupération des valeurs manquantes de l'échantillon filtré
missing_values = df_filtered.isna().sum().sort_values(ascending=False)  

# Affichage de la somme des valeurs manquantes par colonne de l'échantillon filtré
missing_values
```

### 3.2°) Filtrage en fonction des valeurs uniques

```python
# Filtrer votre échantillon avec une certaine tolérance concernant les valeurs uniques par colonne
# Exclure les colonnes avec une seules valeur unique
nunique = df.nunique()
df_filtered = df_filtered.loc[:, nunique > 1]
df_filtered
```

```python
# Nom des colonnes respectant notre seuil de tolérance àl 'égard des valeurs manquantes
df_filtered.columns
```

```python
# Boucle pour récupérer chaque colonne de type integer pour les transformer en colonne de type datetime 
for c in df_filtered.columns:
    if 'datetime' in str(c):
        df_filtered[c] = pd.to_datetime(df_filtered[c])
```

# 4°) Filtrage des Lignes (Réduction de la taille)

### 4.1°) Filtrage en fonction des valeurs manquantes

```python
# nombre de colonnes de l'échantillon filtré
columns_number = df_filtered.shape[1]
```

```python
# Filtrer votre échantillon avec une certaine tolérance concernant les valeurs non-nulles par ligne
# Exemple
# Eliminer les lignes où +70% des valeurs sont nulles (NaN)
df_filtered = df_filtered.dropna(thresh=int(0.3 * columns_number)) # Seuil minimal de 30% de valeurs non-nulles
df_filtered
```

### 4.2°) Filtrage en fonction d'une colonne donnée et d'une valeur donnée

```python
## Se focaliser sur certaines colonnes
# df_filtered = df_filtered[df_filtered["nom_de_la_colonne"].str.contains("valeur_de_la_colonne", na=False)]
## Exemple avec la colonne labels et la valeur Point Vert, Fabriqué en France
df_filtered_focus = df_filtered[df_filtered["labels"].str.contains("Point Vert, Fabriqué en France", na=False)]
df_filtered_focus
```
